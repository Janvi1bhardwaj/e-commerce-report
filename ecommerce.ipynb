```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_excel("ecommerce.xlsx")
df.head(2)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.fillna({"Product Base Margin": df["Product Base Margin"].mean()}, inplace=True)
```

```python
df.isnull().sum()
```

# Order Priority

```python
df["Order Priority"].value_counts()
```

```python
df["Order Priority"].unique()
```

```python
df["Order Priority"]=df["Order Priority"].replace('Critical ','Critical')
```

```python
df["Order Priority"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Order Priority",data=df)
plt.title("Count Of Order Priority")
plt.savefig("Count Of Order Priority.jpg")
plt.show()
```

# Ship Mode

```python
df["Ship Mode"].value_counts()
```

```python
x=df["Ship Mode"].value_counts().index
y=df["Ship Mode"].value_counts().values
```

```python
plt.figure(figsize=(6,5))
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.title("Ship Mode",fontsize=15)
plt.legend(loc=2)
plt.savefig("Ship Mode.jpg")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x="Ship Mode",data=df,hue="Product Category")
plt.savefig("Ship Mode with pc.jpg")
plt.show()
```

# Customer Segment

```python
plt.figure(figsize=(6,3))
sns.countplot(x="Customer Segment",data=df)
plt.savefig("Customer Segment.jpg")
plt.show()
```
